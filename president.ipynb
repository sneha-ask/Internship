```python
import requests
import re
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
soup = bs(page.text)
soup
```

```python
title= soup.find('ul', class_='listing cf')
title
```

```python
# first we will use html tag where we have the first title
first_title = soup.find('div',class_='presidentListing')
print(first_title)
```

```python
first_title.text
```

```python
first_title.text.replace('\n','')
```

```python
President_name =[] # empty list for store the titles

for i in soup.find_all('div',class_='presidentListing'):
    a=i.text
    b=re.findall(r"(.*)\s\(",a)
    President_name.append(b[0])
    
President_name
```

```python
soup.find_all('div',class_='presidentListing')
```

```python
office =[] # empty list for store the titles

for i in soup.find_all('div',class_='presidentListing'):
    a=i.text
    b=re.findall(r"Term of Office:(.*)",a) #using re expression, all the informtion after Term of Office: is extracted
    c=b[0].strip() #here as there are extra spaces bedore 25, to clean the data strip method is been used.
    office.append(c)
    
office
```

```python
df = pd.DataFrame({'Former President':President_name,'Term of Office':office})
df
```
