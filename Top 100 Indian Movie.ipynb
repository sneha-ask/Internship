```python
import requests
import re
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page
```

```python
soup = bs(page.text)
soup
```

```python
title= soup.find('tbody', class_='lister-list')
title
```

```python
# first we will use html tag where we have the first title
first_title = soup.find('td',class_='titleColumn')
print(first_title)
```

```python
first_title.text
```

```python
first_title.text.replace('\n','')
```

```python
rate = soup.find('td',class_='ratingColumn')
rate.text
rate.text.replace('\n','')
```

```python
yearOR = soup.find('span',class_='secondaryInfo')
yearOR.text
```

```python
#now we have all the tags in which there are the movie titles 
# now we will extract the text from these tags one bye one by looping over these tags 

movie_names =[] # empty list for store the titles

for i in soup.find_all('td',class_='titleColumn'):
    movie_names.append(i.text.replace('\n',''))
    
movie_names
```

```python
# scraping all rating

Rating= [] # empty list

for i in soup.find_all('td',class_='ratingColumn imdbRating'):
    Rating.append(i.text.replace('\n',''))
    
Rating
```

```python
#scraping year of release

Year = [] # empty list

for i in soup.find_all('span',class_='secondaryInfo'):
    Year.append(i.text.split('|')[0])
        
    
Year
```

```python
df = pd.DataFrame({'Movie name':movie_names,'RATING':Rating,'year of release':Year})
df
```

```python
#using PANDAS SQL
df1=df[['Movie name','RATING','year of release']].head(100)
df1
```
